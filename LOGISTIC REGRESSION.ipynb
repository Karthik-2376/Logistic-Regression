```python
#Reading of csv File

import numpy as np
import pandas as pd

df = pd.read_csv("diabetes2.csv")
```

```python
# taking of X and Y values

X = df.drop("Outcome", axis=1).values
Y = df["Outcome"].values.reshape(-1, 1)
```

```python
#splittinng training and testing data

indices = np.arange(len(X))
split = int(0.6 * len(X))

X_train = X[:split]
X_test  = X[split:]
Y_train = Y[:split]
Y_test  = Y[split:]


mean = X_train.mean(axis=0)
std  = X_train.std(axis=0)

X_train = (X_train - mean) / (std + 1e-8)
X_test  = (X_test - mean) / (std + 1e-8)

# adding ones to 1st column
ones = np.ones((X_train.shape[0], 1))
X_train = np.hstack((ones, X_train))
```

```python
#sigmoid
def sigmoid(z):
   return 1 / (1 + np.exp(-z))
```

```python
#weigths gradient descent cost accuracy
theta = np.zeros((X_train.shape[1], 1))
learning_rate = 0.01
epochs = 2000
m = len(Y_train)
print("_________________________________________\n")

for i in range(epochs):
    h = sigmoid(X_train @ theta)
    gradient = (1/m) * (X_train.T @ (h - Y_train))
    theta -= learning_rate * gradient

    if i % 200 == 0:
        cost = -(1/m) * np.sum(
            Y_train*np.log(h + 1e-8) + (1-Y_train)*np.log(1-h + 1e-8)
        )
        print(f"Epoch {i:4d}   | Cost: {cost:10.6f}")



print("_________________________________________")

X_test  = np.hstack((np.ones((X_test.shape[0], 1)), X_test))
Y_prob = (sigmoid(X_test @ theta) >= 0.5).astype(int)
accuracy = np.mean(Y_prob == Y_test) * 100

print("\nWeigths : \n",theta)

print("_________________________________________")
print("\nTraining size : ",len(X_train))
print("Test size     : ",len(X_test))
print("_________________________________________")

print("\nModel Accuracy:", accuracy, "%")


```

```python
def confusion_matrix(Y_test,Y_pred):
  TP = np.sum((Y_test == 1) & (Y_pred == 1))
  FN = np.sum((Y_test == 1) & (Y_pred == 0))
  FP = np.sum((Y_test == 0) & (Y_pred == 1))
  TN = np.sum((Y_test == 0) & (Y_pred == 0))
  return TP,FN,FP,TN

TP, FN, FP, TN = confusion_matrix(Y_test, Y_prob)

print("\nConfusion Matrix")
print("----------------")
print(f"TP : {TP}    FP : {FP}")
print(f"FN : {FN}    TN : {TN}")


accuracy  = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP + 1e-8)
recall    = TP / (TP + FN + 1e-8)
f1_score  = 2 * precision * recall / (precision + recall + 1e-8)

print("\nModel Evaluation Metrics")
print("-------------------------")
print(f"Accuracy  : {accuracy*100:.2f}%")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1_score:.4f}")

```
